## COMPLETEZZA POKEMON

Più che considerare la completezza a livello di intera tabella, credo sia sensato valutare colonna per colonna, dal momento che per alcuni attributi la presenza di un NaN indica che quel dato non esiste per quella specifica entità, e non che è un dato mancante (per esempio, Bulbasaur ha un NaN nella colonna 'Evolves_from' solo perchè non evolve da nessuno, non perchè manchi quella informazione).

```python
pokemon = pd.read_csv('Dataset integrati 2.0/df_pokemon.csv')
pokemon
```

```python
pkmn_column_nan = pd.DataFrame(pokemon.isna().sum(), columns=['Total NaN'])
pkmn_column_nan
```

```python
#Calcolo la percentuale di NaN per singola colonna, considerando che abbiamo 1098 righe
pkmn_column_nan['% NaN'] = (pkmn_column_nan['Total NaN']/1098)*100
pkmn_column_nan = pkmn_column_nan.round(decimals=2)
pkmn_column_nan
```

Sulle colonne relative alle statistiche, oltre a 'Rarity' e 'Has_gender_diff' abbiamo un 2.91% di NaN, che fanno però riferimento alle 32 varianti Gigantamax presenti nel dataset. Ci mancano quindi le informazioni di queste specifiche varianti.

Per quanto riguarda le colonne 'Variant', 'Evolves_from' e 'Type2' l'elevata percentuale di NaN non indica una mancanza di informazione, quanto più il fatto che quasi l'80% dei Pokemon non hanno varianti, circa il 50% sono Pokemon base che non evolvono da nessun altro Pokemon e poco meno del 47% dei Pokemon non hanno un secondo tipo (infatti la colonna 'Type1' non presenta NaN).

Non abbiamo informazioni per l'attributo 'VGC2022_rules' per quasi il 35% dei Pokemon. C'è poi il 68.85% di NaN nelle colonne 'Monthly Usage (k)', 'Usage Percent (%)' e 'Monthly Rank': questo dato, però, non è particolarmente preoccupante, poichè i Pokemon che per la colonna 'VGC2022_rules' hanno valore 'Banned' o un NaN non possono avere informazioni per queste tre colonne. Allo stesso modo, la presenza di NaN per gli altri Pokemon denota semplicemente come tali Pokemon non sia utilizzati in competitivo.

```python
## Completezza intero dataset
print(pokemon.size)
print(pokemon.isna().sum().sum())
print(pokemon.isna().sum().sum() / pokemon.size * 100)
```

# ACCURATEZZA

Dimensione di qualità che valuta quanto un valore è vicino alla rappresentazione reale dell'oggetto che rappresenta (es. quanto il nome del Pokemon è vicino al nome reale del Pokemon stesso). Può essere valutata, a livello sintattico, in termini di Edit Distance.

### Pokemon

```python
from nltk.metrics import edit_distance
```

```python
pokemon_api_names = pokedex_api[['ID', 'Name']]
pokemon_api_names.head()
```

```python
pokemon_scrap_names = pokedex_scraping.groupby('#', as_index=False)['Name'].first()
pokemon_scrap_names.columns = ['#', 'Reference']
pokemon_scrap_names.head()
```

```python
pokemon_names = pd.merge(pokemon_api_names, pokemon_scrap_names, left_on='ID', right_on='#')
pokemon_names.drop('#', axis=1, inplace=True)
pokemon_names
```

```python
pokemon_names["Edit Distance"] = pokemon_names.loc[:, ["Name", "Reference"]].apply(lambda x: edit_distance(*x), axis=1)
pokemon_names[pokemon_names["Edit Distance"] != 0]
```

Sui Pokemon è stato possibile fare analisi dell'Edit Distance poichè abbiamo l'ID del Pokedex che individua univocamente ogni singolo Pokemon. Su items, mosse, ecc... non so bene come fare.
