## COMPLETEZZA POKEMON

Più che considerare la completezza a livello di intera tabella, credo sia sensato valutare colonna per colonna, dal momento che per alcuni attributi la presenza di un NaN indica che quel dato non esiste per quella specifica entità, e non che è un dato mancante (per esempio, Bulbasaur ha un NaN nella colonna 'Evolves_from' solo perchè non evolve da nessuno, non perchè manchi quella informazione).

```python
pokemon = pd.read_csv('Dataset integrati 2.0/df_pokemon.csv')
pokemon
```

```python
pkmn_column_nan = pd.DataFrame(pokemon.isna().sum(), columns=['Total NaN'])
pkmn_column_nan
```

```python
#Calcolo la percentuale di NaN per singola colonna, considerando che abbiamo 1098 righe
pkmn_column_nan['% NaN'] = (pkmn_column_nan['Total NaN']/1098)*100
pkmn_column_nan = pkmn_column_nan.round(decimals=2)
pkmn_column_nan
```

Sulle colonne relative alle statistiche, oltre a 'Rarity' e 'Has_gender_diff' abbiamo un 2.91% di NaN, che fanno però riferimento alle 32 varianti Gigantamax presenti nel dataset. Ci mancano quindi le informazioni di queste specifiche varianti.

Per quanto riguarda le colonne 'Variant', 'Evolves_from' e 'Type2' l'elevata percentuale di NaN non indica una mancanza di informazione, quanto più il fatto che quasi l'80% dei Pokemon non hanno varianti, circa il 50% sono Pokemon base che non evolvono da nessun altro Pokemon e poco meno del 47% dei Pokemon non hanno un secondo tipo (infatti la colonna 'Type1' non presenta NaN).

Non abbiamo informazioni per l'attributo 'VGC2022_rules' per quasi il 35% dei Pokemon. C'è poi il 68.85% di NaN nelle colonne 'Monthly Usage (k)', 'Usage Percent (%)' e 'Monthly Rank': questo dato, però, non è particolarmente preoccupante, poichè i Pokemon che per la colonna 'VGC2022_rules' hanno valore 'Banned' o un NaN non possono avere informazioni per queste tre colonne. Allo stesso modo, la presenza di NaN per gli altri Pokemon denota semplicemente come tali Pokemon non sia utilizzati in competitivo.

```python
## Completezza intero dataset
print(pokemon.size)
print(pokemon.isna().sum().sum())
print(pokemon.isna().sum().sum() / pokemon.size * 100)
```

## COMPLETEZZA MOVES

```python
moves = pd.read_csv('Dataset integrati 2.0/df_moves.csv')
moves
```

```python
moves_column_nan = pd.DataFrame(moves.isna().sum(), columns=['Total NaN'])
moves_column_nan
```

```python
#Calcolo la percentuale di NaN per singola colonna, considerando che abbiamo 821 righe
moves_column_nan['% NaN'] = (moves_column_nan['Total NaN']/821)*100
moves_column_nan = moves_column_nan.round(decimals=2)
moves_column_nan
```

Per circa il 4% delle mosse non abbiamo informazioni sull'attributo 'Effect' e per un altro 4% non abbiamo informazione riguardo l'attributo 'Damage_class': nel primo caso è plausibile che siano mosse senza un effetto secondario (la mossa 'Aqua Tail', per esempio, è una mossa che infligge solamente danno, senza applicare alcun effetto aggiuntivo (https://bulbapedia.bulbagarden.net/wiki/Aqua_Tail_(move))), nel secondo caso invece si tratta delle mosse G-Max che non hanno una categoria specifica di danno poichè la mossa può essere sia fisica che speciale a seconda della mossa da cui parte (https://bulbapedia.bulbagarden.net/wiki/Max_Move).

Il 72% delle mosse ha valore mancante per l'attributo 'Prob. (%)': questo però dipende dal fatto che solo alcune mosse hanno degli effetti (come può essere uno status o un drop di una statistica) che si attivano con una certa probabilità, mentre gli altri effetti sono peculiari della mossa e si attivano sempre semplicemente eseguendola.

Per quanto riguarda 'Power' la presenza dei NaN è dettata dalla presenza delle mosse status che non infliggono danni, mentre per la colonna 'Acc.' la presenza dei NaN è conseguenza del fatto che alcune mosse non possono fallire (le mosse per cui 'Acc.'=100 sono mosse che, di base, non falliscono mai, ma in presenza di un drop della precisione di un Pokemon in battaglia possono effettivamente fallire, mentre quelle con valore NaN non vengono influenzate da eventuali drop e non falliscono mai).

Abbiamo poi informazioni mancanti riguardo i PP di 20 mosse.

```python
## Completezza intero dataset
print(moves.size)
print(moves.isna().sum().sum())
print(moves.isna().sum().sum() / moves.size * 100)
```
